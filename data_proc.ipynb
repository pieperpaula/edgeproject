```python
import pandas as pd
```

```python
game_diff = pd.read_csv("game_difference.csv")
```

```python
game_diff
```

```python
game_copy = game_diff.copy()
```

```python
game_copy
```

```python
for ind, row in game_copy.iterrows():
    if game_copy.loc[ind,"SCORE_DIFF"] < 0:
        game_copy.loc[ind,"upset"] = 1
    else:
        game_copy.loc[ind,"upset"] = 0
        

```

```python
for ind,row in game_copy.iterrows():
    if game_copy.loc[ind,"WSEED"] < game_copy.loc[ind,"LSEED"]:
        game_copy.loc[ind,"HSEED"] = game_copy.loc[ind,"WSEED"]
        game_copy.loc[ind,"LSEED"] = game_copy.loc[ind,"LSEED"]
        game_copy.loc[ind,"HTEAM"] = game_copy.loc[ind,"WTEAM"]
        game_copy.loc[ind,"LTEAM"] = game_copy.loc[ind,"LTEAM"]
        game_copy.loc[ind,"HSCORE"] = game_copy.loc[ind,"WSCORE"]
        game_copy.loc[ind,"LSCORE"] = game_copy.loc[ind,"LSCORE"]
    elif game_copy.loc[ind,"LSEED"] < game_copy.loc[ind,"WSEED"]:
        game_copy.loc[ind,"HSEED"] = game_copy.loc[ind,"LSEED"]
        game_copy.loc[ind,"LSEED"] = game_copy.loc[ind,"WSEED"]
        game_copy.loc[ind,"HTEAM"] = game_copy.loc[ind,"LTEAM"]
        game_copy.loc[ind,"LTEAM"] = game_copy.loc[ind,"WTEAM"]
        game_copy.loc[ind,"HSCORE"] = game_copy.loc[ind,"LSCORE"]
        game_copy.loc[ind,"LSCORE"] = game_copy.loc[ind,"HSCORE"]
```

```python
game_copy
```

```python
import numpy as np
```

```python
game_copy = game_copy.drop(["WSEED","WTEAM","WSCORE","Unnamed: 0"], axis=1)
```

```python
game_copy.to_csv('upset_data.csv')
```
